```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch


BASE_MODEL = "Qwen/Qwen2.5-1.5B-Instruct"
CKPT_PATH  = "./checkpoints"

tokenizer = AutoTokenizer.from_pretrained(CKPT_PATH)

base = AutoModelForCausalLM.from_pretrained(BASE_MODEL, torch_dtype=torch.bfloat16, device_map="auto")
model = PeftModel.from_pretrained(base, CKPT_PATH)
model.eval()
print("Model loaded!")

```

```python
def ask(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=1024*3,
            temperature=0.7,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )
    return tokenizer.decode(output[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

response = ask("Problem: What is 25 * 4 + 10? and what is the GCD of the answer and 22")
print(response)
```

```python
response = ask("Problem: Find all real solutions to x^3 - 6x^2 + 11x - 6 = 0")
print(response)
```

```python
response = ask("""Problem: Let $f(x) = x^2 + bx + c$ where $b$ and $c$ are integers.
If $f(f(1)) = f(f(2))$ and $f(1) \\neq f(2)$, find $b + c$.""")
print(response)
```
